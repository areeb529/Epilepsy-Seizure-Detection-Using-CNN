```python
import numpy as  np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('EpilepticSeizure.csv')
df
```

```python
df=df.drop('Unnamed',axis='columns')
df
```

```python
dict={1:0,2:1,3:1,4:2,5:2}
df.y=df.y.map(dict)
df.y
```

```python
df
```

```python
df.y.value_counts()
```

```python
x=df.drop('y',axis='columns')
y=df.y
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=scale.fit_transform(x)
```

```python
x.max()
```

```python
x.min()
```

```python
import keras
```

```python
from keras.utils import to_categorical
y=to_categorical(y)
y
```

```python
y.max()
```

```python
y.min()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=10,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import numpy as np
x_train=x_train.reshape(8050,178,1)
x_test=x_test.reshape(3450,178,1)
x_train.shape , x_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
cnn=keras.Sequential([
    #LAYER 1
    keras.layers.Conv1D(filters=32 , kernel_size=3, activation='relu', input_shape=(178,1)),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPooling1D(2),
    
    
    #LAYER 2
    keras.layers.Conv1D(filters=64 , kernel_size=3, activation='relu'),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPooling1D(2),
    
    #LAYER 3
    keras.layers.Conv1D(filters=128 , kernel_size=3, activation='relu'),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPooling1D(2),

    keras.layers.Flatten(),
    
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(3,activation='softmax')
])
```

```python
cnn.summary()
```

```python
cnn.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
hist=cnn.fit(x_train, y_train , epochs=30,validation_split=0.2,verbose=1)
```

```python
cnn.evaluate(x_test,y_test)
```

```python
y_pred=cnn.predict(x_test)
```

```python
y_pred
```

```python
y_test_class=[np.argmax(val) for val in y_test]
y_pred_class=[np.argmax(val) for val in y_pred]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_test_class,y_pred_class)
print(cm)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(cm,annot=True,fmt='d')
plt.plot()
```

```python
print(classification_report(y_test_class,y_pred_class))
```
